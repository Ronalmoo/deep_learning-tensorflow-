```python
import numpy as np 
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
%matplotlib inline
import zipfile
```

```python
DATA_IN_PATH = '/Users/seungmoo/Library/Mobile Documents/com~apple~CloudDocs/텐서플로와 머신러닝으로 시작하는 자연어처리/tensorflow-ml-nlp-master/4.TEXT_CLASSIFICATION/data_in/'
```

```python
file_list = ['labeledTrainData.tsv.zip', 'unlabeledTrainData.tsv.zip', 'testData.tsv.zip']

for file in file_list:
    zipRef = zipfile.ZipFile(DATA_IN_PATH + file, 'r')
    zipRef.extractall(DATA_IN_PATH)
    zipRef.close()
```

```python
file_list
```

### 데이터 크기

```python
print('파일 크기 : ')
for file in os.listdir(DATA_IN_PATH):
    if 'tsv' in file and 'zip' not in file:
        print(file.ljust(30) + str(round(os.path.getsize(DATA_IN_PATH + file) / 1000000, 2)) +' MB')
        
```

```python
train_data = pd.read_csv(DATA_IN_PATH + 'labeledTrainData.tsv', header = 0, delimiter = '\t', quoting = 3)
train_data.head()
```

### 데이터 개수

```python
print('전체 학습데이터의 개수: {}'.format(len(train_data)))
```

```python
train_data.shape
```

### 각 리뷰의 문자 길이 분포

```python
train_length = train_data['review'].apply(len)
```

```python
train_length.head()
```

```python
plt.figure(figsize=(12, 5))
# histogram
# bins: 히스토그램 값들에 대한 버켓 범위
# range: x축 값의 범위
# alpha: 그래프 색상 투명도
# color: 그래프 색상
# label: 그래프에 대한 라벨
plt.hist(train_length, bins=200, alpha=0.5, color='r', label='word')
plt.yscale('log', nonposy='clip')
plt.title('Log-Histogram of length of review')
# 그래프 x축 라벨
plt.xlabel('Length of review')
# 그래프 y축 라벨
plt.ylabel('Number of review')
```

```python
print('리뷰 길이 최대값: {}'.format(np.max(train_length)))
print('리뷰 길이 최소값: {}'.format(np.min(train_length)))
print('리뷰 길이 평균값: {}'.format(np.mean(train_length)))
print('리뷰 길이 표준편차: {:.2f}'.format(np.std(train_length)))
print('리뷰 길이 중간값: {}'.format(np.median(train_length)))
# 사분위에 대한 경우는 0 ~ 100 스케일로 되어있음 
print('리뷰 길이 제 1 사분위: {}'.format(np.percentile(train_length, 25)))
print('리뷰 길이 제 3 사분위: {}'.format(np.percentile(train_length, 75)))
```

```python
plt.figure(figsize=(12, 5))
# 박스플롯 생성
# 첫번쨰 파라미터: 여러 분포에 대한 데이터 리스트를 입력
# labels: 입력한 데이터에 대한 라벨
# showmeans: 평균값을 마크함

plt.boxplot(train_length,
            labels=['counts'],
            showmeans=True)
```

### 많이 사용된 단어

```python
from wordcloud import WordCloud
cloud = WordCloud(width=800, height=600).generate(' '.join(train_data['review']))
plt.figure(figsize=(20, 15))
plt.imshow(cloud)
plt.axis('off')
```

### 긍, 부정 데이터의 분포

```python
fig, axe = plt.subplots(ncols=1)
fig.set_size_inches(6, 3)
sns.countplot(train_data['sentiment'])
```

```python
print('긍정 리뷰 갯수: {}'.format(len(train_data[train_data['sentiment'] == 0])))
print('부정 리뷰 갯수: {}'.format(train_data['sentiment'].value_counts()[1]))
# print('부정 리뷰 갯수: {}'.format(train_data['sentiment'].value_counts()[]))

```

### 각 리뷰의 단어 개수 분포

```python
train_word_counts = train_data['review'].apply(lambda x: len(x.split(' ')))
```

```python
plt.figure(figsize=(15, 10))
plt.hist(train_word_counts, bins=50, facecolor='r', label='train')
plt.title('Log-Histogram of word count in review', fontsize=15)
plt.yscale('log', nonposy='clip')
plt.legend()
plt.xlabel('Number of words', fontsize=15)
plt.ylabel('Number of reviews', fontsize=15)
```
